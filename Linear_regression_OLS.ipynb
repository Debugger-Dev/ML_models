```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets and models/Datasets/advertising.csv")
df.head()
```

```python
df["total_amount"]=df["TV"]+df["Radio"]+df["Sales"]
df.head()
```

```python
sns.scatterplot(data=df,x="total_amount",y="Sales")
```

```python
sns.regplot(data=df,x="total_amount",y="Sales",color="Green")  #OLS
```

```python
import plotly.express as px
fig = px.scatter(x=df["total_amount"],y=df["Sales"],trendline="ols")
fig.show()
```

```python
x=df["total_amount"]
y=df["Sales"]
m=np.polyfit(x,y,deg=1)[0]
c=np.polyfit(x,y,deg=1)[1]
#returns m and c
```

```python
potential_sales=np.linspace(0,500,100)
predicted_sales=m*potential_sales + c
```

```python
sns.scatterplot(x="total_amount",y="Sales",data=df)
plt.plot(potential_sales,predicted_sales,color="Blue")
```

```python
spend=300
predicted_sales = 0.04868788*spend + 4.24302822
predicted_sales
```

```python
#Y = B3x**3 + B2x**2 + B1x + B0
np.polyfit(x,y,deg=3)   #with 3 DOF
```

```python
potential_sales=np.linspace(0,500,100)
predicted_sales=4.48924833e-07*potential_sales**3 + (-2.92697222e-04)*potential_sales**2 + 1.08893723e-01 *potential_sales + 2.52059799e+00
predicted_sales
```

```python
sns.scatterplot(x="total_amount",y="Sales",data=df)
plt.plot(potential_sales,predicted_sales,color="Green")
```
